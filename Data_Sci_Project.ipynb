```python
import requests
import pandas as pd

API_KEY = "f65f34fd14bfc3e7a64ba44c2f2d94c4663bf2fd"
years = [2018, 2019, 2020, 2021, 2022, 2023]
all_data = []

for year in years:
    url = f"https://api.census.gov/data/timeseries/poverty/histpov2?get=PCTPOV,RACE&time={year}&RACE=12&for=us:1&key={API_KEY}"
    response = requests.get(url)
    if response.status_code == 200:
        data = response.json()
        columns = data[0]
        rows = data[1:]
        all_data.extend(rows)  # Add the rows to our list
    else:
        print(f"Error fetching data for {year}: {response.status_code}")

# Convert to DataFrame
df = pd.DataFrame(all_data, columns=columns)
print(df.head())  # First 5 rows
```

```python
import matplotlib.pyplot as plt

# Make sure PCTPOV is numeric
df['PCTPOV'] = pd.to_numeric(df['PCTPOV'])
df['time'] = pd.to_numeric(df['time'])  # Year as numeric

# Sort by year just in case
df = df.sort_values('time')

# Plot line graph
plt.figure(figsize=(8,5))
plt.plot(df['time'], df['PCTPOV'], marker='o', linestyle='-')
plt.title('Poverty Percentage for Race 1 (White alone) in the US')
plt.xlabel('Year')
plt.ylabel('Poverty Percentage')
plt.grid(True)
plt.show()
```
